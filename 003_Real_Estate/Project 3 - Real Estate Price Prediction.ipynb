##### Откроем локальную или веб-версию файла:
1. напишем функцию поиска файла на локальном диске, 
2. после того как файл найден, прочтем его: 
    - разделитель - знак табуляции, 
    - назначим индексом столбец, содержащий время размещения объявления - 'first_day_exposition', 
    - переведем индекс в формат даты. 

```python
def find(name, path):
    '''
    Function to find the data file stored locally. 
    Takes name of the file and root directory. 
    Scanes the local directory tree for the first orrurance of the file. 
    '''
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)


online_path = 'datasets'
offline_path = '/home/'
file = 'real_estate_data.csv'

try:
    df = (pd.read_csv(f'/{online_path}/{file}',
                      sep='\t',
                      index_col='first_day_exposition',
                      parse_dates=True,
                      dayfirst=True))
except FileNotFoundError:
    df = (pd.read_csv(find(file, offline_path),
                      sep='\t',
                      index_col='first_day_exposition',
                      parse_dates=True,
                      dayfirst=True))
    print("Файл найден на локальном диске.")
else:
    print(f"Файл открыт в режиме online.")
finally:
    print()
    print(
        f"Размер датафрейма: \nколичество строк - {df.shape[0]}, \nколичество столбцов - {df.shape[1]}."
    )
    print()
    print(
        'Развернутая сводка по наименованию столбцов, типу данных, ненулевых значениях и объему памяти:'
    )
    df.info(memory_usage='deep')
```

###### Выводы:
1. пропуски не имеют ярко выраженной закономерности
2. в значениях столбца также присутствуют значения значительно ниже 2,20 и выше 3,5 метров
3. при этом количество тех и других позволяет использовать среднее значение (2.77 м) для замены пропусков
4. т.к. в пятиэтажках и девятиэтажках, как правило, потолки ниже среднего значения, то при заполнении пропусков используем значение первого персентиля (2.52 м)
5. чтобы устранить всплески значений, заменим все значения выше 4,5 метров на значение третьего персентиля(2.80 м).

```python
'''напишем функцию для замены пропусков в зависимости от этажности дома и корректировки высоты потолков 
в зависимости от текущих значений:
'''


def ceiling_height_correction(row):
    current_hight = row['ceiling_height']
    floors = row['floors_total']

    if current_hight > 4.5:
        row['ceiling_height'] = 2.8  # заменяем значения выше 4,5 метра
    elif (np.isnan(current_hight) and
          (floors == 5 or floors == 9)) or (current_hight < 2.2):
        row['ceiling_height'] = 2.52  # заполняем пропуски в пятиэтажках и исправляем значения менее 2,2 метра (СНИП)
    elif (np.isnan(current_hight) and (floors != 5 or floors != 9)):
        row['ceiling_height'] = 2.77  # заполняем пропуски в остальных домах

    return row
```

```python
df = df.apply(ceiling_height_correction, axis=1)
```

```python
# проверим наличие пропусков:
df.ceiling_height.isnull().sum()
```

```python
# построим график распределения квартир по высоте потолков после изменений:
ceiling_plot = df.ceiling_height.hist(bins=50, figsize=(15, 7))
ceiling_plot.set_title(
    'Распределение высоты потолков квартир после обновления данных')
ceiling_plot.set_xlabel('Высота потолков, м')
ceiling_plot.set_ylabel('Количество объектов недвижимости, шт.')
```

Только дом с 34-мя этажами не имеет медианного значения. Вероятно, это единственный дом в выборке. 
Рассмотрим его значение отдельно несколько позже. 

Создадим функцию для заполнения пропусков в зависимости от этажности дома:  

```python
def kitchen_nan_filler(row):
    for x in median_kitchens_by_total_floors.index:
        if row['floors_total'] == x:
            if pd.isnull(row['kitchen_area']):
                row['kitchen_area'] = median_kitchens_by_total_floors[x]
    return row
```

- living_area - данные о жилой площади квартиры

Найдем соотношение общей площади к жилой в зависимости от этажности дома:

```python
groupped_area = df.groupby('floors_total').agg({
    'living_area': 'mean',
    'total_area': 'mean'
})
groupped_area["ratio"] = groupped_area.living_area / groupped_area.total_area
display(groupped_area)
```

Заменим пропуски жилой площади с использованием полученного коэффициента: 

```python
def living_area_nan_filler(row):
    for x in groupped_area.index:
        if row['floors_total'] == x:
            if pd.isnull(row['living_area']):
                row['living_area'] = round(
                    row['total_area'] * groupped_area.loc[x, 'ratio'], 2)
    return row
```

```python
# применим функцию: 
df = df.apply(living_area_nan_filler, axis=1)
```

```python
# проверим полученный результат: 
df.living_area.isnull().sum()
```

```python
# построим график распределения квартир по жилой площади после изменений:
living_area_plot = df.living_area.hist(bins=50, figsize=(15, 7))
living_area_plot.set_title(
    'Распределение квартир по жилой площади после обновления данных')
living_area_plot.set_xlabel('Жилая площадь, м кв.')
living_area_plot.set_ylabel('Количество объектов недвижимости, шт.')
```

Разобьем задачу на 2 этапа: 
1. добавим слово 'город' во все ячейки, начинающиеся с заглавной буквы, 
2. разделим столбец на два: 
- тип населенного пункта, 
- наименование населенного пункта, 
3. Заменим 'поселок' на 'посёлок'

```python
# добавим идентификатор "город" для городов в столбце 'locality_name'
df['locality_name'] = df.apply(
    lambda x: ('город ' + x['locality_name'])
    if x['locality_name'][0].isupper() else x['locality_name'],
    axis=1)
```

```python
# разделим столбец 'locality_name' на столбец типа населенного пункта и его наименование:
df[['locality',
    'just_locality_name']] = df['locality_name'].str.split(r'\s(?=[А-Я])',
                                                           n=1,
                                                           expand=True)
```

```python
# проверим уникальные значения двух новых столбцов:
print(np.sort(df['locality'].unique()))
```

```python
print(np.sort(df['just_locality_name'].unique()))
```

```python
# заменим 'поселок' на 'посёлок' в столбце 'locality'
df['locality'] = df.apply(
    lambda x: x['locality'].replace('поселок', 'посёлок')
    if 'поселок' in x['locality'].lower() else x['locality'],
    axis=1)
```

```python
# т.к. минимальное расстояние не может быть равно 0, то округлим полученные значения в сторону большего и 
# переведём тип данных столбца в цельночисленные значения
spb_df['center_distance'] = spb_df['center_distance'].apply(
    np.ceil).astype(int)
```

```python
# посчитаем среднюю цену для каждого километра и сохраним ее в новую переменную
avg_km_price = spb_df.pivot_table(index='center_distance', values='last_price')
```

```python
# построим график зависимости средней цены от удаленности от центра
avg_km_price.plot(
    kind='bar',
    grid=True,
    legend=False,
    figsize=(15, 5),
    xlabel='Расстояние от центра, км.',
    ylabel='Средняя стоимость недвижимости, млн. руб.',
    title='Зависимость средней цены от расстояния до центра Санкт-Петербурга')
```

**Вывод:**<br>Условной границей центра Санкт-Петербурга является 8-ой километр от центра города. Начиная с 9-го километра средняя стоимость значительно падает.

## Общий вывод

Максимально точно определить рыночную стоимость объектов недвижимости возможно только при условии наличия максимально полной информации об объекте: 
<ul> 
<li> общая площадь,</li>
<li> жилая площадь,</li>
<li> площадь кухни,</li>
<li> этаж квартиры,</li> 
<li> количество этажей в доме,</li> 
<li> расстояние от центра,</li> 
<li> является ли квартира апартаментом или нет.</ul> 
В виду этого данные поля должны быть обязательными при заполнении объявления. При этом формат полей, в которые необходимо вносить данные о площади или расстоянии, необходимо определить, как числовые, а поля, относящиеся к этажности или принадлежности к аппартаментам, должны иметь вид выпадающего списка.<br>
Кроме того, для расчета необходимо уточнять такие расчетные параметры как, стоимость квадратного метра, соотношение общей и жилой площади, соотношение площади кухни к общей площади.<br>
Дополнительными параметрами оценки недвижимости также являются данные о наличии, типе и количестве санузлов, которые к сожалению, не были представлены в данном датасете.
