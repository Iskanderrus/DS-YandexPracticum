```python
def find(name, path):
    '''
    Function to find the data file stored locally. 
    Takes name of the file and root directory. 
    Scanes the local directory tree for the first orrurance of the file. 
    '''
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)


online_path = 'datasets'
offline_path = '/home/'
file = 'games.csv'

try:
    df = (pd.read_csv(f'/{online_path}/{file}'))
except FileNotFoundError:
    df = (pd.read_csv(find(file, offline_path)))
    print("Файл найден на локальном диске.")
else:
    print(f"Файл открыт в режиме online.")
finally:
    print()
    print(
        f"Размер датафрейма: \nколичество строк - {df.shape[0]}, \nколичество столбцов - {df.shape[1]}."
    )
    print()
    print(
        'Развернутая сводка по наименованию столбцов, типу данных, ненулевых значениях и объему памяти:'
    )
    df.info(memory_usage='deep')
```

Так как часть наименований содержит данные о годе выпуска в виде четырехзначного числа, напишем функцию, которая использует данные о годе выпуска из названия и вернет их в ячейку столбца **year_of_release**.

```python
def year_extractor(row):
    extraction = str(row['name']).strip().split()[-1]
    '''убираем возможные лишние пробелы из названия, разделяем строку по пробелам и забираем последнее 
    значение из списка'''
    if len(extraction) == 4:  # проверяем длину полученной строки, если равна 4, обрабатываем далее
        if (extraction.isdigit()) and (pd.isnull(row['year_of_release'])):
            '''если строка может быть переведена в целое число и значение столбца года выпуска 
            имеет пропуск, заменяем год выпуска на нашу переменную'''
            row['year_of_release'] = int(extraction)
    return row
```

```python
# применим полученную функцию к датафрейму:
df = df.apply(year_extractor, axis=1)
```

```python
# проверим полученный результат:
print("Количество оставшихся пропусков в столбце year_of_release:",
      df[df['year_of_release'].isnull()].shape[0])
```

## **Проверка гипотез**

### Гипотеза о равенстве средних пользовательских рейтингов платформ XOne и PC 

Создадим два набора данных для платформ. 

```python
sample_1 = df.query('platform == "XOne"')['user_score']
```

```python
sample_1 = sample_1.dropna()
```

```python
sample_1 = sample_1[sample_1 != 0]
```

```python
sample_2 = df.query('platform == "PC"')['user_score']
```

```python
sample_2 = sample_2.dropna()
```

```python
sample_2 = sample_2[sample_2 != 0]
```

```python
print(
    f'Количество игр на платформе XOne: {len(sample_1)}, \nКоличество игр на платформе PC: {len(sample_2)}.'
)
```

```python
# рассмотрим графическое распределение оценок пользователей: 
figure(figsize=(12, 6), dpi=80)
pyplot.grid(color='black', linestyle="dotted", linewidth=1)
pyplot.hist(sample_1, bins=20, alpha=0.5, label='Платформа XOne')
pyplot.hist(sample_2, bins=20, alpha=0.4, label='Платформа PC')
pyplot.title(label='Распределение оценок пользователей для платформ XOne и PC',
             fontdict={
                 'fontsize': 'x-large',
                 'fontweight': 'semibold'
             })

pyplot.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
pyplot.ylabel("Количество игр", fontsize='large')
pyplot.xlabel("Оценка пользователей", fontsize='large')
```

**Нулевая гипотеза**: средние оценки пользователей платформ XOne и PC за последние 5 лет одинаковы. <br>
**Альтернативная гипотеза**: средние оценки пользователей платформ XOne и PC за последние 5 лет различаются.

```python
alpha = .05  # критический уровень статистической значимости, если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(sample_1, sample_2)

print('p-значение: ', round(results.pvalue, 10))

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
actions = actions[actions != 0]
```

```python
sports = df.query('genre == "Sports"')['user_score']
```

```python
sports = sports.dropna()
```

```python
sports = sports[sports != 0]
```

```python
print(
    f'Количество игр жанра "Action": {len(actions)}, \nКоличество игр жанра "Sports": {len(sports)}.'
)
```

```python
# рассмотрим графическое распределение оценок пользователей: 
figure(figsize=(12, 6), dpi=80)
pyplot.grid(color='black', linestyle="dotted", linewidth=1)
pyplot.hist(actions, bins=20, alpha=0.5, label='Жанр "Actions"')
pyplot.hist(sports, bins=20, alpha=0.4, label='Жанр "Sports"')
pyplot.title(label='Распределение оценок пользователей для жанров "Actions" и "Sports"',
             fontdict={
                 'fontsize': 'x-large',
                 'fontweight': 'semibold'
             })

pyplot.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
pyplot.ylabel("Количество игр", fontsize='large')
pyplot.xlabel("Оценка пользователей", fontsize='large')
```

**Нулевая гипотеза**: средние оценки пользователей жанров **Action** и **Sports** за последние 5 лет одинаковы. <br>
**Альтернативная гипотеза**: средние оценки пользователей жанров **Action** и **Sports** за последние 5 лет различаются.

```python
alpha = .05  # критический уровень статистической значимости, если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(
    actions,
    sports,
    equal_var=False  # т.к. размеры выборок существенно отличаются
)

print('p-значение: ', round(results.pvalue, 10))

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
